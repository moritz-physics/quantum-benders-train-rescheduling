```python
# Some example plots

from qcedule.experiments.data import load_results

qres0 = load_results("results/5_8_11_14_const_QUANT.pkl")
qres1 = load_results("results/6_9_12_15_const_QUANT.pkl")
qres2 = load_results("results/7_10_13_const_QUANT.pkl")
qres3 = load_results("results/0-1-2-3-4_const_QUANT.pkl")
sres = load_results("results/up_to_13th_constr_SIM.pkl")
cres = load_results("results/presolve_res.pkl")
bres = load_results("results/greedy_and_gurobi.pkl")
sres2 = load_results("results/simulation.pkl")
```

```python
for r in sres:
    r.method = "simulator"
for r in sres2:
    r.method = "simulator"
```

```python
from qcedule.experiments.exp_framework import plot_qubits

qres = qres0 + qres1

plot_qubits(qres)
```

```python
from qcedule.experiments.exp_framework import plot_metric

plot_metric(
    qres1 + qres0 + qres2 + qres3 + sres, metric="time", ignore_method=[], minmax=True
)
```

```python
from qcedule.experiments.data import add_relative_delay
from qcedule.experiments.exp_framework import plot_metric

qres = qres0 + qres1 + qres2 + qres3 + sres + cres + bres + sres2
add_relative_delay(qres)
plot_metric(
    qres, "time", ignore_diff=[10, 11, 12, 13], ignore_method=["quantum"], minmax=False
)
plot_metric(
    qres,
    "max_qubits",
    ignore_diff=[],
    ignore_method=["greedy", "gurobi", "centralized"],
    minmax=True,
)
plot_metric(
    qres, "iter_num", ignore_diff=[], ignore_method=["centralized"], minmax=False
)
plot_metric(
    qres, "total_dev", ignore_diff=[], ignore_method=["greedy", "gurobi"], minmax=False
)
plot_metric(
    qres,
    "total_dev",
    ignore_diff=[],
    ignore_method=["simulator", "quantum"],
    minmax=False,
)
plot_metric(
    qres,
    "rel_dev",
    ignore_diff=[0, 1, 2, 3, 4],
    ignore_method=["centralized"],
    minmax=False,
    area=False,
)
```
