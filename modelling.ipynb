```python
# Notebook to test and debug modelling

from qcedule.io_utils.file_parser import parse_network_file, parse_train_file
from qcedule.io_utils.translator import build_constraints
from qcedule.routing import benders_algorithm

net = "data/network_OEOEB.txt"
trains = "data/train_OEOEB.txt"
ts = parse_train_file(trains)
G = parse_network_file(net)
PLATFORMS = [21, 6, 0, 33, 24, 7, 27, 14, 30, 16, 17, 2]
```

```python
constr = build_constraints(G, ts)
```

```python
from datetime import timedelta


def get_time(sec):
    t = timedelta(seconds=sec)
    return str(t)


risks = {
    1: [[6, 21], [33, 0], [24, 7], [27, 14], [16, 30]],
    2: [[6, 21], [33, 0], [24, 7], [27, 14], [16, 30], [17]],
    4: [[27, 14], [16, 30], [17]],
    3: [[6, 21], [33, 0], [24, 7], [27, 14], [16, 30]],
    5: [[6, 21], [33, 0], [24, 7], [27, 14], [16, 30]],
    6: [[27, 14], [24, 7], [6, 21], [33, 0]],
}
```

```python
res = benders_algorithm(
    constr, platforms=risks, scp_strat="quantum", depth=2, steps=15, order=True
)
for i, _ in enumerate(ts, start=1):
    for p in plist:
        for (x, y), value in res.items():
            if x == i and y in p and any(y in p for p in risks[i]):
                print((x, y), get_time(value))
        print()
```
